```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings                                                             
warnings.filterwarnings('ignore')
```

```python
cars_df  = pd.read_csv("C:\\Users\\Gayatri\\Downloads\\cars\\cars\\cars.csv")
cars_df
```

```python
cars_df.columns
```

# cars that have mpg greater than 25

```python
mpg_df=cars_df[cars_df['mpg']>25]
result = mpg_df[['Carrs', 'mpg']]
print(result)
```

# List of cars that have more than 6 cyl and have hp greater than 150.

```python
cylinder_hp=cars_df[(cars_df['cyl']>6) & (cars_df['hp']>150)]
print(cylinder_hp[['Carrs', 'cyl', 'hp']])
```

# Total the number of cylinders and the average horsepower.

```python
df=cars_df.groupby('cyl')['hp'].mean().reset_index
df
```

# Total number of cars for each number of gears.

```python
gear_df=cars_df.groupby('gear').size().reset_index(name='total cars')
gear_df
```

# Top 5 cars with the highest mpg.

```python
data = cars_df.sort_values(by='mpg', ascending=False).head(5)
print(data[['Carrs', 'mpg']])
```

# Top 5 heaviest cars.

```python
top_5_weight=cars_df.sort_values(by='wt',ascending=False).head(5)
print(top_5_weight[['Carrs','wt']])
```

# List of cars that have a mpg greater than 20 and are either 6 or 8 cylinders.

```python
listed=cars_df[(cars_df['mpg']>20) & ((cars_df['cyl']==6)|(cars_df['cyl']==8))]
print(listed[['Carrs','mpg','cyl']])
```

# The correlation between horsepower and weight of the cars.

```python
correlation = cars_df['hp'].corr(cars_df['wt'])
print("correlation between hp and wt:",correlation)
```

# cars that have more carb than the average number of carb.


```python
avg_carb= cars_df['carb'].mean()
above_avg_carb = cars_df[cars_df['carb'] > avg_carb]
print(above_avg_carb[['Carrs', 'carb']])
```

# Cars with a cylinder count that is not a standard value.

```python
standard_cylinders = {'4', '6', '8'}
s=list(filter(lambda x: str(x) not in standard_cylinders, cars_df['cyl']))
s
```
